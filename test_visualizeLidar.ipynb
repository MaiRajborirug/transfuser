```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import os

# Create a sample (n, 4) array
n = 400
folder_path = '/media/haoming/970EVO/Pharuj/cdc_eval/240912_tfcbf_noise0.0_rep1_1'
point_cloud = np.load(os.path.join(folder_path,'lidar_'+str(n)+'.npy'))

# Split the array into X, Y, Z, and intensity
point_cloud = point_cloud[(point_cloud[:, 2] > -2.1) & (point_cloud[:, 2] < 0)]
X = point_cloud[:, 0]
Y = point_cloud[:, 1]
Z = point_cloud[:, 2]
intensity = point_cloud[:, 3]  # This is your intensity value

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter with intensity as color
scatter = ax.scatter(X, Y, Z, c=intensity, cmap='viridis', alpha=0.8, s=0.05)

# Add a color bar to show intensity values
cbar = fig.colorbar(scatter, ax=ax)
cbar.set_label('Intensity')

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set the view angle
# Syntax: view_init(elev, azim)
# Top-down view (elev = 90, azim = -90 for viewing X-Y plane)
ax.view_init(elev=90, azim=-90)  # Elevation: 90 degrees, Azimuth: -90 degrees

plt.show()
```
